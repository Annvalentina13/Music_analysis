Univariative Analysis

```python
# Plot 1: Simple bar chart of top 5 artists
plt.figure(figsize=(10, 6))
top_5 = df['artist(s)_name'].value_counts()[:5]
plt.bar(top_5.index, top_5.values, color='skyblue')
plt.xlabel('Artist Name')
plt.ylabel('Number of Tracks')
plt.title('Top 5 Artists by Number of Tracks')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# Plot 2: Simple histogram of streams
plt.figure(figsize=(10, 6))
plt.hist(df['streams'], bins=30, edgecolor='black', color='lightgreen')
plt.title('Distribution of Streams')
plt.xlabel('Number of Streams')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Plot 3: Top 5 artists by total tracks 
plt.figure(figsize=(10, 6))
bottom_5_artists = df['artist(s)_name'].value_counts().tail(5)
plt.bar(bottom_5_artists.index, bottom_5_artists.values, color='salmon')
plt.xlabel('Artist Name')
plt.ylabel('Number of Tracks')
plt.title('Bottom 5 Artists by Number of Tracks')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
# Simple boxplot for top 5 artists
plt.figure(figsize=(10, 6))

# Get top 5 artists by total streams with proper numeric handling
# Convert streams to numeric and handle any non-numeric values
df['streams'] = pd.to_numeric(df['streams'], errors='coerce')

# Group and sort to get top 5
top_5_artists = df.groupby('artist(s)_name')['streams'].sum().sort_values(ascending=False).head(5)

# Prepare data for boxplot
data = [df[df['artist(s)_name'] == artist]['streams'].dropna() for artist in top_5_artists.index]

# Create boxplot with customization
box = plt.boxplot(data, 
                 labels=top_5_artists.index,
                 patch_artist=True)

# Customize colors
for patch in box['boxes']:
    patch.set_facecolor('lightblue')

# Basic formatting
plt.xticks(rotation=45, ha='right')
plt.xlabel('Artist Name')
plt.ylabel('Number of Streams')
plt.title('Stream Distribution for Top 5 Artists')
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Create 3D scatter plot of top artists' streams vs tracks vs popularity
from mpl_toolkits.mplot3d import Axes3D

# Create figure and 3D axes
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Get top 15 artists data
top_artists = df.groupby('artist(s)_name').agg({
    'streams': 'sum',
    'track_name': 'count'
}).sort_values('streams', ascending=False).head(15)

# Create scatter plot
scatter = ax.scatter(top_artists['streams'], 
                    top_artists['track_name'],
                    range(len(top_artists)),
                    c=top_artists['streams'],
                    cmap='viridis',
                    s=100)

# Customize axes
ax.set_xlabel('Total Streams')
ax.set_ylabel('Number of Tracks')
ax.set_zlabel('Artist Rank')

# Add colorbar
plt.colorbar(scatter, label='Total Streams')

# Add title
plt.title('3D Visualization of Top 15 Artists: Streams vs Tracks vs Rank')

# Rotate the plot for better view
ax.view_init(elev=20, azim=45)

# Add artist labels
for i, artist in enumerate(top_artists.index):
    ax.text(top_artists['streams'][i], 
            top_artists['track_name'][i], 
            i,
            artist[:10] + '...' if len(artist) > 10 else artist)

plt.tight_layout()
plt.show()
```

```python
# Create a figure with 2x2 subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))

# 1. Horizontal bar chart of top 10 tracks by streams
top_10_tracks = df.nlargest(10, 'streams')[['track_name', 'streams']]
ax1.barh(top_10_tracks['track_name'], top_10_tracks['streams'], color='skyblue')
ax1.set_title('Top 10 Tracks by Streams')
ax1.set_xlabel('Number of Streams')

# 2. Pie chart of top 5 artists by total streams
top_5_streams = df.groupby('artist(s)_name')['streams'].sum().nlargest(5)
ax2.pie(top_5_streams.values, labels=top_5_streams.index, autopct='%1.1f%%', 
        colors=plt.cm.Set3(np.linspace(0, 1, 5)))
ax2.set_title('Stream Share of Top 5 Artists')

# 3. Line plot of streams distribution (sorted)
streams_sorted = np.sort(df['streams'])
ax3.plot(streams_sorted, color='green', linewidth=2)
ax3.set_title('Stream Distribution (Sorted)')
ax3.set_xlabel('Track Rank')
ax3.set_ylabel('Number of Streams')
ax3.grid(True, alpha=0.3)

# 4. Bar plot of track count by year
df['year'] = pd.to_numeric(df['released_year'], errors='coerce')
year_counts = df['year'].value_counts().sort_index()
ax4.bar(year_counts.index, year_counts.values, color='salmon')
ax4.set_title('Number of Tracks by Year')
ax4.set_xlabel('Year')
ax4.set_ylabel('Number of Tracks')
ax4.tick_params(axis='x', rotation=45)

# Adjust layout
plt.tight_layout()
plt.show()
```
